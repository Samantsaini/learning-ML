```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#linespace is a function in numpy that return evenly spaced 100 value b/w 0 to 10
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(y,z)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'y--')
plt.subplot(2,2,2)
plt.plot(x,z,'g-*')
plt.subplot(2,2,3)
plt.plot(x,y,'b+-')
```

```python
x=np.linspace(20,40,15)
y=3*x+5
plt.plot(x,y)
plt.xlabel('Angles')
plt.ylabel('Sine Wave')
plt.title('Sine Wave')
plt.show()
```

```python
x=np.linspace(20,40,15)
y=3*x+5
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Linear Graph')
plt.show()
```

```python
x=np.linspace(20,40,15)
y=81-x*x
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Linear Graph')
plt.show()
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Parabola')
plt.show()
```
